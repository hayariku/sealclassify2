```python
learn_inf = load_learner('export.pkl')
```

```python
btn_upload = widgets.FileUpload(layout=Layout(width='auto', height='auto'))
# layout=Layout(width='auto', height='auto')
b1 = widgets.Button(description='Classify', text_color="red", layout=Layout(width='auto', height='auto'))
# layout=Layout(width='auto', height='auto')
b1.style.keys
# b1.style.button_color = "blue"

btn_run = b1
out_pl = widgets.Output()
out_pl.clear_output()
lbl_pred = widgets.Label()

out_pl.add_class("left-space-margin")

def on_click_classify(change):
    for i in hi:
       i.value = " "
    img = PILImage.create([f.content.tobytes() for f in btn_upload.value][-1])
    out_pl.clear_output()
    with out_pl: display(img.to_thumb(500,500))
    pred_class,pred_idx,probabilities = learn_inf.predict(img)
    for i in range(len(pred_class)):
      hi[i].value = nameProbDisplay(pred_class[i], probabilities[i])
    if len(pred_class) == 0:
      hi[0].value = "No seal or sea lion detected."

btn_run.on_click(on_click_classify)

def create_expanded_button(description, button_style):
    return Button(description=description, button_style=button_style, layout=Layout(height='auto', width='auto'))


# header_button = create_expanded_button('Header', 'success')
# left_button = create_expanded_button('Left', 'info')
# center_button = create_expanded_button('Center', 'warning')
# right_button = create_expanded_button('Right', 'info')
# footer_button = create_expanded_button('Footer', 'success') 


#inst_label = widgets.HTML(
    #value="Hello <b>World</b>",
    #placeholder='Some HTML',
    #description='Some HTML',
#)

#widgets.Label('Upload a photo of a seal or a sea lion!', text_align='center')
  
#html = 
top_label = widgets.HTML(
    value= "<style> p{text-align: center; color: navy;} .myheader{background-color: lightblue; text-align: right;} </style> <div class='myheader'> <p> Seal or sea lion classifier </p> </div>"
)



test_button = widgets.Button(description="test", layout =Layout(height="auto", width="auto"))

#filler = widgets.Button(description="test", layout =Layout(height="auto", width="auto"))
```
